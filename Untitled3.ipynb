```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/OIP (1).jpg")
```

```python
cv2_imshow(img)
```

```python
import numpy as np
```

```python
k1=np.array(([[1,0,1],[1,0,1],[1,0,1]]))
```

```python
img2=cv2.filter2D(src=img,ddepth=-1,kernel=k1)
```

```python
img3=cv2.medianBlur(img,3)
```

```python
cv2_imshow(img3)
```

```python
img4=cv2.medianBlur(img,7)
```

```python
cv2_imshow(img4)
```

```python
img5=cv2.min(img,5)
```

```python
cv2_imshow(img5)
```

```python
img6=cv2.max(img,5)
```

```python
cv2_imshow(img6)
```
